```python
import pathlib
import requests
from lxml import etree

html = requests.get("https://p5.readthedocs.io/en/latest/reference/index.html").text
tree = etree.HTML(html)

urls = tree.xpath('//li/a/@href')
urls = ['https://p5.readthedocs.io/en/latest/reference/' + i for i in urls if i.endswith('.html') and not '..' in i]
urls
```

```python
import time
import markdown2opml
import html2markdown_clipboard as h2mc

path = pathlib.Path.cwd()

for i, url in enumerate(urls):
    html = requests.get(url).text
    markdown = h2mc.html2markdown(html)
    opml = markdown2opml.md2opml(markdown)
    with open(path / f"test/output_{i}.opml", "w", encoding='utf-8') as f:
        f.write(opml)

```
